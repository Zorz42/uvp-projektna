```python
import pandas as pd
import matplotlib.pyplot as plt

problems = pd.read_csv("data/cses_problems.csv")
problems.head(10)
```

```python
sorted_df = problems.sort_values(by="solves", ascending=False)
sorted_df.head(10)
```

Pogledamo najbolj rešene probleme. Tukaj dobimo prvih šest problemov in presenetljivo že dva trivialna problema iz sklopa iz dinamičnega programiranja.


```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(sorted_df)), sorted_df["solves"], alpha=0.5)
plt.yscale("log")
plt.ylabel("Število rešitev (logaritemsko)")
plt.grid(True)
plt.show()
```

Graf nam razkrije, da število rešitev dejansko eksponentno pade.

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(problems)), problems["solves"], alpha=0.5)
plt.yscale("log")
plt.ylabel("Število rešitev (logaritemsko)")
plt.grid(True)
plt.show()
```

Tukaj se vidi, da so problemi načeloma težji proti koncu, ampak samo na zelo grobo.

```python
# Least solved problems
sorted_df = problems.sort_values(by="solves", ascending=True)
sorted_df.head(10)
```

Tukaj dobimo najslabše rešene probleme. Vidimo, da so večinoma težki problemi, ni pa to popolna metrika težavnosti, saj na to vpliva tudi starost problema (novejši problemi so manj rešeni). Na to vpliva tudi to, da so nekateri problemi bolj sitni za implementacijo, kar lahko odvrne ljudi od reševanja. Obstaja tudi nekaj problemov, ki se pojavijo na spletnih straneh, ki razlagajo algoritme in podatkovne strukture, kar lahko poveča število rešitev, recimo v "USACO Guide" je veliko problemov iz CSES. Primer lahkega problema je "Bouncing Ball Cycle", ki je nov, zelo nadležen za implementacijo in nima posebno težkega razmisleka, npr. "Two Stacks Sorting" ima bistveno težji razmislek, pa vendar ga sploh ni na v top 10 najmanj rešenih problemov.

```python
sorted_df = problems.sort_values(by="solve_ratio", ascending=False)
sorted_df.head(10)
```

Tukaj imamo probleme z najvišjim razmerjem (št. pravilnih rešitev) / (št. posizkusov). Rešitve in poizkusi se štejejo enkrat na uporabnika, torej če imam npr. 10 napačnih poizkusov, se bo število poizkusov povečalo zgolj za 1. Če samo preberem nalogo in razmislim, bo število poizkusov ostalo enako. Na tej tabeli na vrhu sploh niso zares vsi problemi trivialni. Npr. "Collecting Numbers Distribution" je precej težek problem. Na to razmerje bolj vpliva sam tip problema. Če ima problem veliko razmerje, potem je verjtno tak, da če ima rešitev napako, jo avtor lahko najde preden sploh odda problem (npr. zelo močen in specifičen sample, taka rešitev, da se avtor lahko hitro prepriča, da je dovolj hitra in pravilna).

```python
plt.figure(figsize=(10, 6))
plt.scatter(range(len(sorted_df)), sorted_df["solve_ratio"], alpha=0.5)
plt.ylabel("Razmerje")
plt.grid(True)
plt.show()
```

Tukaj se vidi, da je pri večina problemov kar visoko razmerje, potem pa pri nekaterih precej pade.

```python
# Problems with lowest solve ratio
sorted_df = problems.sort_values(by="solve_ratio", ascending=True)
sorted_df.head(10)
```
